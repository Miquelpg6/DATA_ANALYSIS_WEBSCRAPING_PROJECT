```python
import requests
from bs4 import BeautifulSoup

# URL base de las jornadas
base_url = "https://www.fcf.cat/resultats/2223/futbol-sala/lliga-primera-divisio-catalana-futbol-sala/bcn-gr-2/jornada-"

# Nombre del equipo a buscar
team_to_find = "SD ESPANYOL B"

# Recorrer las jornadas de la 1 a la 26
for jornada in range(1, 27):
    url = f"{base_url}{jornada}"
    
    # Enviar una solicitud HTTP GET a la URL
    response = requests.get(url)
    
    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Analizar el contenido HTML de la página
        soup = BeautifulSoup(response.text, 'lxml')

        # Buscar el resultado para "SD ESPANYOL B" en esta jornada
        result = None

        # Supongamos que los nombres de los equipos y los resultados están en elementos HTML específicos
        matches = soup.find_all('table', class_='uppercase w-100 fs-12_tp fs-11_ml table_resultats')  # Reemplaza 'match-class' por la clase real

        for match in matches:
            # Extraer los nombres de los equipos y los resultados (ajusta según la estructura HTML)
            team1 = match.find('td', {"class": 'p-5 resultats-w-equip tr'}).text.strip()  # Reemplaza 'team1-class' por la clase correcta
            team2 = match.find('td', {"class": 'p-5 resultats-w-equip tl'}).text.strip()  # Reemplaza 'team2-class' por la clase correcta
            score = match.find('td', {"class": 'p-5 resultats-w-resultat tc'}).text.strip().replace("ACTA TANCADA", "")  # Reemplaza 'score-class' por la clase correcta

            # Verificar si el partido involucra a "SD ESPANYOL B" (o la representación correcta)
            if team_to_find in team1 or team_to_find in team2:
                result = f"{team1} {score} {team2}"
                break  # Salir del bucle si se encuentra el resultado

        if result:
            print(f"Jornada {jornada}: {result}")
        else:
            print(f"SD ESPANYOL B no se encontró en los resultados de la jornada {jornada}")
    else:
        print(f"Fallo al recuperar la página de la jornada {jornada}. Código de estado: {response.status_code}")
```

```python


# The URL of the website
url = "https://www.fcf.cat/calendari/2223/futbol-sala/lliga-primera-divisio-catalana-futbol-sala/bcn-gr-2"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'lxml')

    # Find all JORNADAS and DATES by inspecting the page's HTML structure
    jornadas_and_dates = []

    # Assuming the JORNADAS and DATES are contained within specific HTML elements
    jornada_elements = soup.find_all('th', {'colspan': '4'})
    date_elements = soup.find_all('th', {'colspan': '3'})

    # Iterate through the elements and collect the information
    for jornada, date in zip(jornada_elements, date_elements):
        jornada_text = jornada.text.strip()
        date_text = date.text.strip()
        jornadas_and_dates.append((jornada_text, date_text))

    # Print the extracted JORNADAS and DATES
    for i, (jornada, date) in enumerate(jornadas_and_dates, start=1):
        print(f"JORNADA {i}: {jornada}")
        print(f"FECHA {i}: {date}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
